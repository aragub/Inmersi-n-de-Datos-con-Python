# Aula 3

Vamos a crear una nueva columna que representa el calculo **"El valor por metro cuadrado"**

```python
inmuebles["Valor_m2_Millon"] = inmuebles ["Precio_Millon"]/inmuebles["Area"]
inmuebles.head(3)
```

Aqui vamos agrupar los datos y los vamos agrupar en esta caso por barrio y vamos a tomar el promedio (En una agrupacion siempre se debe utilizar una funció de agregación)

```python
inmuebles.groupby("Barrio").mean()
```

Vamos a agrupar por barrio pero en vez de aplicar directamente la funcion "mean" vamos a usar la función "sum" que es para sumar.

```python
datos_barrio = inmuebles.groupby("Barrio").sum()
datos_barrio
```

Vamos a calcular el promedio con base a la informacion anterior.

```python
datos_barrio["Valor_m2_Barrio"] = datos_barrio["Precio_Millon"]/datos_barrio["Area"]
datos_barrio
```

Aqui creamos un diccionario de datos de nuestro datos "Barrio" el Valor de precio por m2.


```python
m2_barrio = dict(datos_barrio["Valor_m2_Barrio"])
```

```python
inmuebles["Valor_m2_Barrio"] = inmuebles["Barrio"]
inmuebles["Valor_m2_Barrio"] = inmuebles["Valor_m2_Barrio"].map(m2_barrio)
inmuebles.head(5)
```

Aqui vamos a ver el top 10 de barrios que contienen más inmuebles a la venta.


```python
top_barrios = inmuebles["Barrio"].value_counts()[:10].index
```

```python
datos_barrio.reset_index(inplace=True)
datos_barrio
```

Podremos realizar consultas 

```python
datos_barrio.query("Barrio in @top_barrios")
```

Vamos a utilizar un    para resolver el desafio 

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(x="Barrio", y="Valor_m2_Barrio", data = datos_barrio.query("Barrio in @top_barrios"))
ax.tick_params(axis="x",rotation=45)
```

Descripcion estadistica

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="Precio_Millon", data = datos_barrio.query("Barrio in @top_barrios"))
ax.tick_params(axis="x",rotation=45)
plt.show()
```
